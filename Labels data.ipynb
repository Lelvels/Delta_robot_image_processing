```python
import shutil
filenames = label_data["FileName"]
cashew_types = label_data["Type"]
index = 0
origin = "data/all_crop_images/crop_images/"
cashew_path = "data/cashew/"
no_cashew_path = "data/no_cashew/"
for filename in filenames:
    filepath = origin + filename
    cashew_type = cashew_types[index]
    print(index, filepath, cashew_type)
    if(cashew_type == 1):
        shutil.copy(filepath, cashew_path+filename)
    else:
        shutil.copy(filepath, no_cashew_path+filename)
    index = index + 1
```
